```python
df['Time Period'] = df['Title'].str.extract(r'\)([A-Za-z]{3} \d{4} - [A-Za-z]{3} \d{4})')
```

```python
# prompt: create a column duration by using time period column.
# Example if  'Apr 2009 - Jul 2010' then the duration column should get the difference

import pandas as pd
from dateutil.relativedelta import relativedelta

def calculate_duration(time_period_str):
    if pd.isna(time_period_str):
        return None
    try:
        start_date_str, end_date_str = time_period_str.split(' - ')
        start_date = pd.to_datetime(start_date_str, format='%b %Y')
        end_date = pd.to_datetime(end_date_str, format='%b %Y')

        # Calculate the difference in years and months
        diff = relativedelta(end_date, start_date)
        years = diff.years
        months = diff.months

        # Represent the duration as a string
        duration_str = f"{years} years, {months} months"
        return duration_str

    except ValueError:
        return None

df['Duration'] = df['Time Period'].apply(calculate_duration)

df
```
